# Task B

```python
# total iterations 
itr = 25

# initializing with 25 so the graphs looks beautiful 
```

**Name**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
from random import choice

fname=('Nabeel','Qasim','Rohan','Jameel','Abbas','Adeel','Imran','Sajjad',"Tanveer")
lname=('Ali','Mustafa','Haris','Tariq','Khurram','Abid','Waqar','Raza','Rana','Shah','Mobeen','Tahir')

name=[]
for i in range(itr):
  n = "".join("%s %s" % (choice(fname), choice(lname)))   
  name.append(n)

print(name)
```

**Age**

```python
age=[]
for i in range(itr):
  x=random.randint(18,22)
  age.append(x)

print(age)
```

**Height**

```python
height=[]
for i in range(itr):
  x=random.randint(60,75)
  height.append(x)

print(height)
```

**GPA**

```python
gpa=[]
for i in range(itr):
  x=round(random.uniform(2.00,4.00),2)
  gpa.append(x)

print(gpa)
```

**Create CSV file**

```python
import csv

f=open("students.csv","a",newline="")
writer=csv.writer(f)
header=("Name","Age","Height","GPA")
writer.writerow(header)

for i in range (itr):
  tup2=(name[i],age[i],height[i],gpa[i])
  writer.writerow(tup2)

f.close()
df=pd.read_csv("students.csv")

import os
os.remove("students.csv")

display(df)
```
