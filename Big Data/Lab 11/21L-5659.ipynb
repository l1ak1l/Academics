```python
from pyspark import SparkContext,SparkConf
conf = SparkConf().setAppName("lab 11").setMaster("local[*]")
sc = SparkContext(conf=conf)
```

```python
lst=[(1,2),(3,4),(3,6),(4,5)]
Rdd = sc.parallelize(lst)
Rdd_Reduced = Rdd.reduceByKey(lambda x, y: x + y)

for pair in Rdd_Reduced.collect():
    print(pair)
```

```python
rdd=sc.textFile("shakespeare.txt")
word_counts = rdd.flatMap(lambda line: line.split()).map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)
for pair in word_counts.collect():
    print(pair)
```
