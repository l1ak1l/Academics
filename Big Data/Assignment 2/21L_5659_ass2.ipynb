```python
x1,y1=make_blobs(n_features=2,n_samples=1500,centers=3 ,shuffle=True,cluster_std=[1.5,3,1],random_state=170)
plt.scatter(x1[:,0],x1[:,1])
plt.show()
```

```python
x2,y2=make_moons(n_samples=6260,noise=0.05)  
plt.scatter(x2[:,0],x2[:,1])
plt.show()
```

```python
x31, y31 = make_circles(n_samples=6260, random_state=3,noise=0.05, factor=0.4)
x32, y32 = make_circles(n_samples=6260, random_state=3,noise=0.05, factor=0.7)
for i in range(0,len(y32)):
    if y32[i]==True:
        y32[i]=2
        
x3=np.concatenate((x31,x32))
y3=np.concatenate((y31,y32))
plt.scatter(x3[:, 0], x3[:, 1])
plt.show()
```

```python
inertias=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x1)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,10),inertias,marker='o')
plt.title("Elbow Method")
plt.show()    

start=time.time()
kmeans=KMeans(n_clusters=3)
kmeans.fit(x1)
end=time.time()
plt.scatter(x1[:,0],x1[:,1],c=kmeans.labels_, cmap='rainbow')
plt.title("KMeans Custering")
plt.show()
print("Time : ",end-start)
print("Purity : ",entropy(y1))
```

```python
inertias=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(x2)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,10),inertias,marker='o')
plt.title("Elbow Method")
plt.show()    

# from elbow method k=2 is best for clustering
start=time.time()
kmeans=KMeans(n_clusters=3)
kmeans.fit(x2)
end=time.time()

plt.scatter(x2[:,0],x2[:,1],c=kmeans.labels_ ,cmap='plasma')
plt.title("KMeans Custering")
plt.show()
print("Time : ",end-start)
print("Purity : ",entropy(y2))
```
